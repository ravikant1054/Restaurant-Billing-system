```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
import os

menu = input("enter hello").lower()

price_list = {
    "samosa": 80,
    "cofee": 50,
    "berger": 80,
    "pizza": 350,
    "momos": 120,
    "chole bhature": 150,
    "dhosa":130,
    "chwomin": 100,
    "gol gappe": 80

}

order_id = int(input("enter your order id"))
file_name = "Restaurant_Data.csv"

if menu == "hello":

    print("""
1.SAMOSA  80
2.COFEE   50
3.BERGER  80
4.PIZZA   350
5.MOMOS   120
6.CHOLE BHATURE   150
7.DHOSA   130
8.CHWOMIN 100
9.GOL GAPPE 80
""")

    order = {}
    n = int(input("how many items do you want: "))

    for i in range(n):
        item = input("items: ").lower()
        qty = int(input("quantity: "))
        order[item] = qty

        hana = input("want to change something yes__no: ").lower()
        if hana == "yes":
            old = input("which item do you want to change: ").lower()
            if old in order:
                new = input("new item (same if qty only): ").lower()
                qty = int(input("new quantity: "))
                del order[old]
                order[new] = qty
            else:
                print("NOT IN ORDER")

    total = 0
    for item, qty in order.items():
        if item in price_list:
            total += price_list[item] * qty
        else:
            print("ITEM IS NOT AVAILBLE")

    print("\nTOTAL WITH GST(18%) =", total * 1.18)
    print("YOUR ORDER =", order)
    print("THANK YOU FOR COMING!")

    
    plt.bar(order.keys(), order.values())
    plt.title("CUSTOMER ORDER SUMMARY")
    plt.xlabel("FOOD ITEMS")
    plt.ylabel("QUANTITY")
    plt.show()

    data = []
    for item, qty in order.items():
        if item in price_list:
            price = price_list[item]
            total_item = price * qty
            data.append([order_id, item, qty, price, total_item])

    df = pd.DataFrame(data, columns=["order_id", "Item", "Quantity", "Price", "total"])

    
    if os.path.exists(file_name):
        df.to_csv(file_name, mode='a', header=False, index=False)
    else:
        df.to_csv(file_name, index=False)

    print(f"\n Data saved in {file_name}")

    
    history = pd.read_csv(file_name)

    customer_data = history[history["order_id"] == order_id]

    if len(customer_data) > 0:
        predicted_item = (
            customer_data
            .groupby("Item")["Quantity"]
            .sum()
            .idxmax()
        )

        print("\n--------------------------------------")
        print("PREDICTION RESULT")
        print("Next time this customer will likely order →", predicted_item)
        print("--------------------------------------")
    else:
        print("No previous data to predict")

else:
    print("please type 'hello' to start ordering")

```
